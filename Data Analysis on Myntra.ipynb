## 3.2 Price Distribution

### 3.2.1  Most Expensive Product (variant_compare_at_price)

```python
df[df['variant_compare_at_price'] == df['variant_compare_at_price'].max()]
```

### 3.2.2 Most Expensive Product (variant_price)

```python
df[df['variant_price'] == df['variant_price'].max()]
```

### 3.2.3 Average discount rate

```python
df['Average discount_rate'] = ((df['variant_compare_at_price'] - df['variant_price']) / df['variant_compare_at_price']) * 100

discounted_products = df[df['variant_compare_at_price'].notnull()]
discounted_products = discounted_products[discounted_products['variant_price'] < discounted_products['variant_compare_at_price']]

average_discount_rate = discounted_products['discount_rate'].mean()
print("Average discount rate:", average_discount_rate)
```

## 3.3 Shortlisting Products based on specific filters

```python
df_pr = df[df['variant_price'] > 2000]
df_pr = df_pr[df_pr['variant_price'] < 3000]
df_pr = df_pr[df_pr['product_type'] == 'Straight Kurta']


df_pr.sort_values(by = 'variant_price', ascending = False).head(10)
```

## 3.4 Finding the top 10 Brands with most number of products

```python
data = []

for brand in df['brand'].unique():
    
    c = 0
    for i in df['brand']:
        if (i == brand):
            c += 1

    data.append([brand, c])
    
data = pd.DataFrame(data, columns = ['name','products'])

data.sort_values(by = 'products', ascending = False).head(10)
```

## 3.5 Finding the top 10 Most Expensive Brands (variant_compare_at_price)

```python
df['variant_compare_at_price'] = pd.to_numeric(df['variant_compare_at_price'], errors='coerce')
```

```python
df.groupby('brand')['variant_compare_at_price'].mean().sort_values(ascending=False).head(10)
```
